```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
file_id='17onQcxDf0tJEIGRP51KdqqSlroDIDZ7D'
url=url = f'https://drive.google.com/uc?export=download&id={file_id}'
df=pd.read_csv(url)

#checking for null values
df.isnull().sum()

# Drop column with many nulls
df.drop(columns=['fielder'],inplace=True)

# Create a 'ball_number' feature
df['ball_number'] = df['over'] * 6 + df['ball']

# Plot settings
plt.figure(figsize=(12, 6))
plt.scatter(df['ball_number'], df['batsman_runs'], alpha=0.5,label='Runs Per Ball',marker='o',s=5,color='darkblue')

#titles and labels
plt.title('IPL: Runs Scored on Each Ball', fontsize=14, fontweight='bold')
plt.xlabel('Ball Number')
plt.ylabel('Runs')

#legend and grid
plt.legend(loc='upper right', title='IPL')
plt.grid(True,color='crimson', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```
